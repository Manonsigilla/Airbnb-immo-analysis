## Étape 2 — Nettoyage des données

### 2.1 Inspection des colonnes suspectes (rating, bedrooms, baths)

```python
# Inspection des colonnes au type suspect
print("=== RATING — valeurs uniques (10 premières) ===")
print(df['rating'].value_counts().head(10))
print(f"\nNombre de valeurs uniques : {df['rating'].nunique()}")

print("\n=== BEDROOMS — valeurs uniques ===")
print(df['bedrooms'].value_counts().head(10))

print("\n=== BATHS — valeurs uniques ===")
print(df['baths'].value_counts().head(10))
```

### 2.2 Plan de nettoyage

```python
print("=== BATHS — valeurs uniques ===")
print(df['baths'].value_counts().head(15))
```

### 2.3 — Conversion de la colonne `rating`
La colonne contient 3 593 fois la valeur textuelle `"No rating"` qui correspond aux logements n'ayant pas encore reçu de note. On les remplace par NaN (valeur manquante propre) puis on convertit la colonne en numérique.

```python
# Remplacement de "No rating" par NaN
df['rating'] = df['rating'].replace('No rating', np.nan)

# Conversion en numérique (errors='coerce' transforme tout texte restant en NaN par sécurité)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Vérification
print(f"Type de la colonne : {df['rating'].dtype}")
print(f"Valeurs manquantes : {df['rating'].isna().sum():,} ({df['rating'].isna().mean()*100:.1f}%)")
print(f"\nStatistiques :")
print(df['rating'].describe())
```

### 2.4 — Conversion de la colonne `bedrooms`
- `"Studio"` → 0 (un studio est, par définition, sans chambre séparée)
- `"Not specified"` (13 lignes) → NaN, puis suppression de ces lignes (négligeable sur 20 758)

```python
# Remplacement des valeurs textuelles
df['bedrooms'] = df['bedrooms'].replace({'Studio': '0', 'Not specified': np.nan})

# Conversion en numérique
df['bedrooms'] = pd.to_numeric(df['bedrooms'], errors='coerce')

# Suppression des lignes où bedrooms est NaN (très peu de lignes)
df = df.dropna(subset=['bedrooms'])

print(f"Type de la colonne : {df['bedrooms'].dtype}")
print(f"Distribution :")
print(df['bedrooms'].value_counts().sort_index())
```

### 2.5 — Conversion de la colonne `baths`
Même traitement que `bedrooms` : `"Not specified"` (13 lignes) → NaN puis suppression. À noter qu'on observe une valeur extrême (11.5 salles de bain) qui sera filtrée à l'étape suivante.

```python
# Remplacement des valeurs textuelles
df['baths'] = df['baths'].replace('Not specified', np.nan)

# Conversion en numérique
df['baths'] = pd.to_numeric(df['baths'], errors='coerce')

# Suppression des lignes où baths est NaN
df = df.dropna(subset=['baths'])

print(f"Type de la colonne : {df['baths'].dtype}")
print(f"Distribution :")
print(df['baths'].value_counts().sort_index())
```

### 2.6 — Suppression des outliers (valeurs aberrantes)
Justification des seuils :
- **`price`** ≤ 1 000 \$/nuit : couvre 99\% du marché new-yorkais, élimine les valeurs extrêmes ( max observé : 100 000 \$ ).
- **`minimum_nights`** ≤ 365 : élimine les locations longue durée déguisées (max observé : 1 250 nuits).
- **`beds`** ≤ 10 : conserve uniquement les vrais Airbnb, exclut les hôtels camouflés.
- **`baths`** ≤ 5 : élimine les erreurs de saisie (max observé : 11.5 salles de bain).

```python
lignes_avant = len(df)

df = df[
    (df['price'] >= 10) & (df['price'] <= 1000) &
    (df['minimum_nights'] <= 365) &
    (df['beds'] <= 10) &
    (df['baths'] <= 6)
]

lignes_apres = len(df)
print(f"Lignes supprimées : {lignes_avant - lignes_apres:,}")
print(f"Lignes restantes : {lignes_apres:,} ({lignes_apres/lignes_avant*100:.1f}% conservé)")
```

### 2.7 Vérification finale

```python
# Reset de l'index après suppressions
df = df.reset_index(drop=True)

print(f"Dataset final : {df.shape[0]:,} lignes × {df.shape[1]} colonnes")
print(f"\nTypes des colonnes critiques :")
print(df[['price', 'rating', 'bedrooms', 'baths', 'beds', 'number_of_reviews']].dtypes)
print(f"\nStatistiques après nettoyage :")
df[['price', 'minimum_nights', 'beds', 'baths', 'rating']].describe()
```

## Étape 3 — Enrichissement (Feature Engineering)
Création de nouvelles variables pertinentes pour l'analyse d'investissement :
1. **`revenu_estime`** : revenu mensuel estimé par logement
2. **`categorie_prix`** : segmentation Low-cost / Standard / Luxe
3. **`type_hote`** : Particulier vs Professionnel (selon le nombre de logements possédés)

### 3.1 — `revenu_estime` : revenu mensuel estimé
Formule : `revenu_estime = price × reviews_per_month`

Cette estimation repose sur l'hypothèse que chaque avis correspond à une réservation. C'est une approximation conservative (en réalité, seuls ~50% des voyageurs laissent un avis), mais elle reste utile pour **comparer les logements entre eux**.

```python
# Création de la colonne revenu estimé mensuel
df['revenu_estime'] = df['price'] * df['reviews_per_month']

# Aperçu
print("Statistiques du revenu estimé mensuel :")
print(df['revenu_estime'].describe().round(2))
```

### 3.2 — `categorie_prix` : segmentation du marché
Découpage en trois segments pour faciliter l'analyse :
- **Low-cost** : < 100 $/nuit
- **Standard** : 100 à 250 $/nuit
- **Luxe** : > 250 $/nuit

Ces seuils correspondent approximativement aux quartiles du marché new-yorkais.

```python
# Création de la catégorie de prix avec pd.cut
bins = [0, 100, 250, df['price'].max() + 1]   # bornes des intervalles
labels = ['Low-cost', 'Standard', 'Luxe']

df['categorie_prix'] = pd.cut(df['price'], bins=bins, labels=labels, right=False)

# Vérification de la répartition
print("Répartition par catégorie de prix :")
print(df['categorie_prix'].value_counts())
print(f"\nEn pourcentage :")
print((df['categorie_prix'].value_counts(normalize=True) * 100).round(1).astype(str) + ' %')
```

### 3.3 — `type_hote` : Particulier vs Professionnel
Variable bonus qui distingue :
- **Particulier** : possède 1 logement (loue probablement sa résidence)
- **Multi-propriétaire** : 2 à 5 logements
- **Professionnel** : 6 logements ou plus (gestion semi-industrielle)

Cette segmentation est utile pour un fonds d'investissement : on observe différemment un marché dominé par des pros vs un marché de particuliers.

```python
# Fonction de catégorisation
def categoriser_hote(n):
    if n == 1:
        return 'Particulier'
    elif n <= 5:
        return 'Multi-propriétaire'
    else:
        return 'Professionnel'

df['type_hote'] = df['calculated_host_listings_count'].apply(categoriser_hote)

# Répartition
print("Répartition par type d'hôte :")
print(df['type_hote'].value_counts())
print(f"\nEn pourcentage :")
print((df['type_hote'].value_counts(normalize=True) * 100).round(1).astype(str) + ' %')
```
